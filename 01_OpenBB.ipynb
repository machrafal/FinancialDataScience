## Historical data for individual expiration

```python
expirations = [
    "2025-12",
    "2026-12",
    "2027-12",
    "2028-12",
    "2029-12",
    "2030-12",
]
```

```python
contracts = []
for expiration in expirations:
    df = (
        obb
        .derivatives
        .futures
        .historical(
            symbol="CL",
            expiration=expiration,
            start_date="2020-01-01",
            end_date="2022-12-31"
        )
    ).rename(columns={
        "close": expiration
    })
    contracts.append(df[expiration])
```

```python
historical = (
    pd
    .DataFrame(contracts)
    .transpose()
    .dropna()
)
```

```python
print(historical)
```

```python
historical.iloc[-1].plot()
```
