## Manipulating and transforming DataFrames

```python
import numpy as np
import pandas as pd
from openbb import obb
obb.user.preferences.output_type = "dataframe"
```

```python
asset = obb.equity.price.historical(
    "AAPL",
    provider="yfinance"
)
benchmark = obb.equity.price.historical(
    "SPY",
    provider="yfinance"
)
```

## Creating new columns

```python
benchmark
```

```python
asset.columns
```

```python
columns = [
    "open",
    "high",
    "low",
    "close",
    "volume",
    "dividends",
]
asset.columns = columns
benchmark.columns = columns
```

```python
asset["price_diff"] = asset.close.diff()
benchmark["price_diff"] = benchmark.close.diff()
```

```python
asset["symbol"] = "AAPL"
benchmark["symbol"] = "SPY"
```

## Joining options data together to create straddle prices

```python
chains = obb.derivatives.options.chains(
    "AAPL", provider="cboe")
```

```python
chains
```

```python
expirations = chains.expiration.unique()
calls = chains[
    (chains.option_type == "call")
    & (chains.expiration == expirations[5])
]
puts = chains[
    (chains.option_type == "put")
    & (chains.expiration == expirations[5])
]
```

```python
calls_strike = calls.set_index("strike")
puts_strike = puts.set_index("strike")
```

```python
joined = calls_strike.join(
    puts_strike,
    how="left",
    lsuffix="_call",
    rsuffix="_put"
)
```

```python
prices = joined[["last_trade_price_call",
                 "last_trade_price_put"]]
```

```python
prices["straddle_prices"] = prices.sum(axis=1)
```

## Grouping by multiple columns

```python
(
    chains
    .groupby(
        ["option_type", "strike", "expiration"]
    )
    .open_interest
    .sum()
)
```

## Applying different methods to different columns

```python
(
    chains
    .groupby(
        ["option_type", "strike", "expiration"]
    )
    .agg({
        "last_trade_price": "max",
        "open_interest": "sum"
    })
)
```

## Applyihng custom functions

```python
(
    chains
    .groupby(["option_type"])
    .apply(lambda x: (x["ask"] - x["bid"]).mean(), include_groups=False)
)
```

## Grouping and transforming data

```python
(
    chains
    .groupby("expiration")
    .last_trade_price
    .transform(lambda x: (x - x.mean()) / x.std())
)
```

## Volatility of returns

```python
import numpy as np
import pandas as pd
from openbb import obb
obb.user.preferences.output_type = "dataframe"
```

```python
df = obb.equity.price.historical(
    "AAPL",
    start_date="2020-01-01",
    provider="yfinance"
)
df
```

```python
close = df["close"]
close
```

```python
returns = close.pct_change()
returns
```

```python
std_dev = returns.dropna().std()
std_dev
```

```python
annualized_std_dev = std_dev * np.sqrt(252)
annualized_std_dev
```

```python
close.index = pd.to_datetime(close.index)
(
    close
    .pct_change(freq="ME")
    .dropna()
    .std()
    * np.sqrt(12)
)
```

```python
(
    close
    .pct_change(freq="QE")
    .dropna()
    .std()
    * np.sqrt(4)
)
```

```python
(
    close
    .pct_change()
    .rolling(window=22)
    .std()
    * np.sqrt(252)
).plot()
```

```python
returns +=1
```

## Custom functions

```python
df.apply(lambda x: x["high"] - x["low"], axis=1)
```

```python
def fcn(row):
    return row["high"] - row["low"]
df.apply(fcn, axis=1)
```

```python
df["valid"] = df.apply(
    lambda x: x["low"] <= x["close"] <= x["high"], axis=1)
df[df.valid == False]
```

```python
def calculate_range(row, high_col, low_col, threshold):
    range = row[high_col] - row[low_col]
    return range if range > threshold else np.nan
```

```python
threshold = 1.5
df["range"] = df.apply(
    calculate_range,
    args=("high", "low", threshold),
    axis=1
)
df.range
```
